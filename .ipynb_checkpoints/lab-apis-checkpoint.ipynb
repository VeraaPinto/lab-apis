```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from spotipy.exceptions import SpotifyException

#Point 1: List  favorite artists
artists = ["MUSE", "IMAGINE DRAGONS", "BILLIE EILISH"]
```

```python
#Point 2: Fetch Top Tracks
# Function to fetch top tracks of an artist
def get_top_tracks(artist_name):
    results = sp.search(q='artist:' + artist_name, type='artist', limit=1)
    artist_id = results['artists']['items'][0]['id']
    top_tracks = sp.artist_top_tracks(artist_id)['tracks']
    top_tracks_list = [track['name'] for track in top_tracks[:5]]
    return top_tracks_list

# Top tracks for each artist in the list
top_tracks_list = [track for artist_name in artists for track in get_top_tracks(artist_name)]
```

```python
#Point 3:Discover Related Artists

# Function to find related artists
def find_related_artists(artist_name):
    try:
        results = sp.search(q='artist:' + artist_name, type='artist', limit=1)
        if not results['artists']['items']:
            print(f"Artist {artist_name} not found.")
            return []
        artist_id = results['artists']['items'][0]['id']
        related_artists = sp.artist_related_artists(artist_id)['artists']
        related_artists_list = [artist['name'] for artist in related_artists[:5]]
        return related_artists_list
    except (IndexError, SpotifyException) as e:
        print(f"Error fetching related artists for {artist_name}: {e}")
        return []
    
# Get related artists
related_artists = [related for artist_name in artists for related in find_related_artists(artist_name)]
related_artists_top_tracks = [track for related_artist in related_artists for track in get_top_tracks(related_artist)]
#function find_related_artists not working because API is deprecated
```

### Getting a Playlist's Details
To fetch details about a specific playlist, you can use the playlist method. You'll need the playlist's Spotify ID.

In this example, we will use the following playlist id: *37i9dQZF1DXd9zR7tdziuQ*

```python
playlist_id = "2Tg6Pz4Ol2pnsshGBjraoL"
playlist = sp.playlist(playlist_id)
```

```python
print(playlist['name'])  # Print the playlist's name
print(playlist['description'])  # Print the playlist's description
```

### Getting Tracks from a Playlist
If you want to get the tracks from a specific playlist, you can use the playlist_tracks method.

```python
tracks = sp.playlist_tracks(playlist_id)
for track in tracks['items']:
    print(track['track']['name'])  # Print each track's name
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:
   - Choose any one of the top 5 featured playlists (you can choose the one you personally find most interesting or simply pick one randomly).
   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:
   - Extract and display the names of the first 10 tracks in the chosen playlist.

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
#point 1 - Extract and display the names and IDs of the top 5 featured playlists
#sp.featured_playlists() method not working because API is deprecated

# Function to fetch featured playlists
def get_featured_playlists(sp, limit=5):
    featured_playlists = sp.featured_playlists(limit=limit)
    return featured_playlists['playlists']['items']

# Fetch top 5 featured playlists
featured_playlists = get_featured_playlists(sp)

# Display the names and IDs of the top 5 featured playlists
for i, playlist in enumerate(featured_playlists, 1):
    print(f"{i}. {playlist['name']} - ID: {playlist['id']}")
```

```python
#point 2 - Fetch and display its name, description, and total track count

# Choose a playlist ID
playlist_id = "2YRe7HRKNRvXdJBp9nXFza"

def get_playlist_details(sp, playlist_id):
    try:
        playlist = sp.playlist(playlist_id)
        name = playlist['name']
        description = playlist['description']
        total_tracks = playlist['tracks']['total']
        return name, description, total_tracks
    except spotipy.exceptions.SpotifyException as e:
        print(f"SpotifyException: {e}")
        return None, None, None

# Fetch playlist details
name, description, total_tracks = get_playlist_details(sp, playlist_id)

# Display the playlist details
if name and description and total_tracks:
    print(f"Name: {name}")
    print(f"Description: {description}")
    print(f"Total Track Count: {total_tracks}")
else:
    print("Failed to fetch playlist details.")
```

```python
#point 3 - Extract and display the names of the first 10 tracks in the chosen playlist

def get_playlist_tracks(sp, playlist_id, limit=10):
    try:
        results = sp.playlist_tracks(playlist_id, limit=limit)
        tracks = results['items']
        return tracks
    except spotipy.exceptions.SpotifyException as e:
        print(f"SpotifyException: {e}")
        return []

# Fetch the first 10 tracks
tracks = get_playlist_tracks(sp, playlist_id, limit=10)

# Display the names of the first 10 tracks
for i, track in enumerate(tracks, 1):
    track_name = track['track']['name']
    print(f"{i}. {track_name}")
```

```python
#point 4 - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track

# Create a dictionary with track names and associated artists
track_artist_dict = {}
for track in tracks:
    track_name = track['track']['name']
    artist_names = [artist['name'] for artist in track['track']['artists']]
    track_artist_dict[track_name] = artist_names

# Display the dictionary
print(track_artist_dict)
```
